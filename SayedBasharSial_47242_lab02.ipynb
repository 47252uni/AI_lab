# Python List Methods Lab Task_02

```python
# 1. append() add elements
fruits = ["Apple", "Mango", "Banana"]
fruits.append("Orange")

print("After append():", fruits)
```

```python
# 2. clear() - Removes all elements
fruits = ["Apple", "Mango", "Banana"]
fruits.clear()

print("After clear():", fruits)
```

```python
# 3. copy() - Returns a copy of the list
fruits = ["Apple", "Mango", "Banana"]
new_list = fruits.copy()

print("Original list:", fruits)
print("Copied list:", new_list)
```

```python
# 4. count() - Returns the number of elements with the specified value
fruits = ["Apple", "Mango", "Banana", "Apple"]

print("Count of Apple:", fruits.count("Apple"))
```

```python
# 5. extend() - Adds elements of another list to the end
fruits = ["Apple", "Mango"]
more_fruits = ["Banana", "Orange"]

fruits.extend(more_fruits)

print("After extend():", fruits)
```

```python
# 6. index() - Returns index of first element with specified value
fruits = ["Apple", "Mango", "Banana"]

print("Index of Mango:", fruits.index("Mango"))
```

```python
# 7. insert() - Adds an element at a specified position
fruits = ["Apple", "Mango", "Banana"]

fruits.insert(1, "Orange")

print("After insert():", fruits)
```

```python
# 8. pop() - Removes element at specified position
fruits = ["Apple", "Mango", "Banana"]

fruits.pop(1)

print("After pop():", fruits)
```

```python
# 9. remove() - Removes item with specified value
fruits = ["Apple", "Mango", "Banana"]

fruits.remove("Mango")

print("After remove():", fruits)
```

```python
# 10. reverse() - Reverses the order of the list
fruits = ["Apple", "Mango", "Banana"]

fruits.reverse()

print("After reverse():", fruits)
```

```python
# 11. sort() - Sorts the list
fruits = ["Banana", "Apple", "Mango"]

fruits.sort()

print("After sort():", fruits)
```

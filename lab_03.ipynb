```python
import numpy as np

d3array = np.array([
    [   # Depth 0
        [1, 2, 3],    # Row 0
        [4, 5, 6],    # Row 1
    ],
    [   # Depth 1
        [7, 8, 9],    # Row 0
        [10, 11, 12], # Row 1
    ],
    [   # Depth 2
        [13, 14, 15], # Row 0
        [16, 17, 18], # Row 1
    ]
])

print("The dimension of array d3array is ", d3array.ndim)
print(d3array)

# d3array[depth_index][row_index][column_index]
item = d3array[2][0][0]  # Depth 2, Row 0, Column 0
print("Item at d3array[2][0][0] is: ", d3array[2][0][0])
```

```python
# Numpy array slicing
import numpy as np

d1 = np.array([6,8,1,2,54,5,3])   # 1-D Array
print("Elements of array d1 are ", d1)

print("Elements slicing 'd1[1:5]' ", d1[1:5])   # 1-D Array slicing "The result includes the start index, excludes the end index"
print("Elements slicing 'd1[:3]' ", d1[:3])     # Slice elements form start index[0] to index[3], index [3] excluded
print("Elements slicing 'd1[3:]' ", d1[3:])     # Slice elements form index[3] to last index
print("Elements slicing 'd1[-3:-1]' ", d1[-3:-1])  # Slice elements (negative slicing) form index[-3] to index[-1], index[-1] excluded

d2 = np.array([[5,9,3,4],[21,65,8,99],[56,98,1,2]])   # 2-D Array
print("Elements of array d2 are ", d2)

print("Elements slicing 'd2[0,1:3]' ", d2[0,1:3])   # 2-D Array slicing : Slice elements form array[0], index[1] to index[3], index[3] excluded
```

```python
import numpy as np

# b as a view of a
a = np.array([1, 2, 3, 4, 5, 6])
print("Array a", a)

b = a[:2]
print("Array b", b)

b += 1
print("Array b after increment if a", b)
print("Array a =", a, "; array b =", b)

# create a new array, using numpy copy array
a = np.array([1, 2, 3, 4])
b = a[:2].copy()
b += 1

print("a =", a, "b =", b)
```
